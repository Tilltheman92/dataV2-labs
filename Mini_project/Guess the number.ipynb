```python
# needed libary
import random

#variables
def restart():
    answer= input("Do you want to play again")
    if answer== "yes":
        play() 
    elif answer=="no":
        print ("Okay, bye bye")
def play():
    secretnum= random.randrange (1,100)
    guess =int(input("Guess a number between 1 and 100:"))
    guess_count=0
    guess_max=7

#guess_again=int(input("Try a different number:"))


    while guess != secretnum and guess_count<guess_max:
        guess_count+=1
        if guess <secretnum:
            print ("The number is higher, try again please")
            guess =int(input("\nGuess a number between 1 and 100:"))
        elif guess >secretnum:
            print ("The number is lower, try again please")
            guess =int(input("\nGuess a number between 1 and 100:"))
            

    if guess==secretnum:
        print ("Congrats you found the number!")
        print ("It took you:",guess_count+1, "guesses!")

    else:
        print ("YOU LOST!You took too many guesses")

play()       
restart()
```
