```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
first=fitbit.groupby(["Months_encoded"])["Steps","Distance"].median()
#1 option
#first.plot()
#2.option
a=list(first.Steps)
b=list(first.index)

f, (ax1,ax2) = plt.subplots(1, 2,figsize=(12.4,12.4))
ax1.bar(b,a)
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
#0=weekend,1=week'
second=fitbit.groupby(["Work_or_Weekend"])["Steps","Distance"].median()
c=list(second.Steps)
d=list(second.index)
ax2.bar(d,c)
f.savefig("Figures/fig.png")
```
