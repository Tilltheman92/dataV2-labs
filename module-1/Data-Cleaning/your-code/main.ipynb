#### 1. Import pandas library.

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
import pymysql
from sqlalchemy import create_engine
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
engine = create_engine('mysql+pymysql://root:Ironhack@localhost/User')
```

#### 4. Import the users table.

```python
user = pd.read_sql_query('SELECT * FROM User.users', engine)
print(user)
```

#### 5. Rename Id column to userId.

```python
user2=user.rename(columns={'Id':'userID'})
user2.head()
```

#### 6. Import the posts table. 

```python
post = pd.read_sql_query('SELECT * FROM User.posts', engine)
post.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
post2=post.rename(columns={'Id':'postID','OwnerUserId':'userID'})
post2.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
newusers= user2[['userID', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
newusers.head()
newposts=post2[['postID', 'Score', 'userID', 'ViewCount', 'CommentCount']]
newposts.head(60)
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged=newusers.merge(newposts)
#inner is default
```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

```python
missing = merged.isnull().sum()
missing[missing > 0]#48396 missing values all in Column ViewCount
```
